```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('creditcard.csv')
dataset.head()
dataset.describe()
#dataset.info()

```

```python
dataset.info()
```

```python
dataset_fraud = dataset[dataset['Class'] == 1]
dataset_non_fraud = dataset[dataset['Class'] == 0]
```

Variables

```python
nb_lines_fraud = len(dataset_fraud)
nb_lines_non_fraud = len(dataset_non_fraud)
nb_lines = len(dataset)
amount_fraud = dataset_fraud['Amount'].sum()
amount_non_fraud = dataset_non_fraud['Amount'].sum()
```

DASHBOARD

```python
print(" === DASHBOARD CREDIT CARD FRAUD === ")
print("Total transactions: ", len(dataset))
print("Fraudulent transactions: ", len(dataset_fraud))
print("Non-fraudulent transactions: ", len(dataset_non_fraud))
print("Total amount of fraudulent transactions: ", amount_fraud)
print("Total amount of non-fraudulent transactions: ", amount_non_fraud)
print("Percentage of fraudulent transactions: ", round((nb_lines_fraud / nb_lines) * 100, 2), "%")
print("Percentage of non-fraudulent transactions: ", round((nb_lines_non_fraud / nb_lines) * 100, 2), "%")
```

Graphs

```python
dataset.corr()['Class'].sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.heatmap(dataset.corr(), cmap='coolwarm', linewidths=
0.5)
plt.title('Correlation Heatmap')
plt.show()
```
